```
ozkim = { "국어" : 90, "'수학" : 80, "과학" : 82, "영어" : 72 }
result =
print(result)
```

<span style="color:red">답변</span>

<span style="font-size:80%">[문제5]<br><br></span>
아래 keys 변수와 vals 변수에 들어있는 값을 이용해 result라는 딕셔너리를 만드는 코드를 작성해주세요

```
keys = ["뉴진스", "아이브", "에스파"]
vals = ["하이브" , "스타쉽", "SM"]
```

<span style="color:red">답변</span>

`````python
keys = ["뉴진스", "아이브", "에스파"];
vals = ["하이브" , "스타쉽", "SM"];

result = dict(zip(keys,vals));

print(result);
`````
