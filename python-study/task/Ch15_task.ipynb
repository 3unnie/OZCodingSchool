<span style="font-size:80%">[문제4]<br><br></span>
```리스트에 저장된 데이터를 아래와 같이 출력하세요```

오피스텔 = [ [201, 202], [301, 302], [401, 402] ]

201호<br>
202호<br>
`----`

301호<br>
302호<br>
`----`

401호<br>
402호<br>
`----`

<span style="color:red">답변</span>

````python
오피스텔 = [ [201, 202], [301, 302], [401, 402] ];


for a in range(len(오피스텔)):
    for b in range(len(오피스텔[a])):
        print(오피스텔[a][b])
    print("----");
````

````python
for 층 in 오피스텔:
    for 호 in 층:
        print(f'{호}호')
    print('---')
````

<span style="font-size:80%">[문제5]<br><br></span>
```ohlc 리스트에는 시가(open), 고가 (high), 저가 (low) , 종가(close)가 날짜별로 저장돼 있다. 종가가 150원보다 큰 경우에만 종가를 출력하는 코드를 작성해주세요```

ohlc = [["open", "high", "low", "close"], [100, 110, 70, 100], [200, 210, 180, 190], [300, 310, 300, 310]]

<span style="color:red">답변</span>

````python
ohlc = [["open", "high", "low", "close"], [100, 110, 70, 100], [200, 210, 180, 190], [300, 310, 300, 310]];

for a in range(len(ohlc)):
    if a > 0:
        if ohlc[a][3] > 150:
            print(a,"번째날 종가:",ohlc[a][3]);
````

````python
for i in ohlc[1:]:
    if i[3] > 150:
        print(i[3]);
````
