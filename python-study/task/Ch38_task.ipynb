requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

```python
import requests

target = requests.get('https://kream.co.kr/')
codeStatus = target.status_code
txt = target.text

print("[상태]:",codeStatus)
print("[HTML 내용 (첫 100자)]:",txt[:100])
```
