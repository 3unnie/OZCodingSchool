<span style="font-size:80%">[문4]<br><br></span>
```문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 만들어주세요```

예 ) make_url("google")

출력 : www.google.com

<span style="color:red">답변</span>

````python
def make_url(str):
    result = "www." + str + ".com"
    return result

name = input("문자열을 입력해주세요")

print(make_url(name))
````
