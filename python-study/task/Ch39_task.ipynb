```python
import datetime,time

def timer_decorator(function):
    def wrapper():
        function()
    return wrapper

@timer_decorator
def timer():
    start= datetime.datetime.now()
    time.sleep(4)
    end= datetime.datetime.now()
    print(f"함수 실행시간 : {end.second-start.second} 초")

timer()
```
