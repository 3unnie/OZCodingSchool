파일을 읽어서 내용을 출력하는 프로그램을 작성하세요.

- 파일이 존재하지 않을 때 발생하는 FileNotFoundError를 처리
- 파일이 존재하지 않을 경우 "파일을 찾을 수 없습니다."라는 메시지를 출력

```python
try:
    f = open("myfile.txt","r")
except FileNotFoundError:
    print("파일을 찾을수 없습니다.")
else:
    line = f.readline()
    while line:
        print(line)
        line = f.readline()
    f.close()
```
