```python
#클래스
class Person:
    def name(self): #self 초기화 의미
        print("조먼지입니다");


hello = Person()

hello.name()
```

```python
#객체와 인스턴스

#같은것 but,
#클래스라는 내용이 동반이 되면 인스턴스
#객체 자체를 언급할때는 객체
```

```python
class Cat:
    def __init__(self): #속성 매직메서드...?
       self.name = "조하루"

    def greeting(self):
        print(self.name)

cat1 = Cat()
cat1.greeting()
```

```python
#self는 인스턴스 자기자신을 의미한다.
```

```python
class Kitty:
    def __init__(self,name,age,sex):
        self.name = name
        self.age = age
        self.sex = sex

    def intro(self):
        print("이름은",self.name,"입니다.")
        print("나이는",self.age,"입니다.")
        print("성별은",self.sex,"입니다.")

cat1 = Kitty("조먼지",0.8,"남")
cat2 = Kitty("조하루",0.2,"남")

cat1.intro()
cat2.intro()

print(cat1.name)
```
