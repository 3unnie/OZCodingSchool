```python
try:
    x = int(input("나눌숫자를 입력하세요"))
    y = x /10
    print (y)
except:
    print("숫자를 올바르게 입력하세요")
else: #try 정상실행후 실행
    print("성공")
finally: #무조건 실행
    print("끝")
```
