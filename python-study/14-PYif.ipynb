```python
#if문

#특정조건일때 실행하는 문

question = int(input("고양이는 총 몇마리 키우시나요?"));

if question == 1:
    #들여쓰기 필수 안하면 error
    print("한마리 키우는군요!");
elif question == 2:
    print("두마리 키우는군요!");
else:
    print("그렇군요!");
```

```python
a = input("a로 시작하는 단어를 입력해주세요");

mid_index = int(len(a)/2);

if a[0] == 'a':
    if len(a) % 2 != 0: #홀수일때
        print(a[mid_index]);
    else: #짝수일때
        print(a[mid_index-1:mid_index+1]);
else:
    print("a로 시작하는 단어를 적어주세요");
```
