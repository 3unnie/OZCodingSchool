```python
#while 반복문

i = 1;
while i < 10:
    print(i,"번째 반복중");
    i += 1;
```

```python
#무수히 많이 실행되어야할때 강점

import random;

i = 0;
while i != 4:
    i = random.randint(1,6);
    print ("랜덤 숫자[",i,"]");
```

```python
먼지 = 10;
while 먼지 > 0:
    print("먼지 간식이",먼지,"개 남았습니다");

    if 먼지 == 3:
        break; # 강제종료
    먼지 -= 1;
else:
    print("먼지간식 없따");
```

```python
a = 0;

while a < 10:
    a += 1;
    if a % 2 == 0:
        continue; # 짝수면 재실행 (위로)
    print (a);
```

```python
a = 0;

while a < 10:
    a += 1;
    if a % 2 != 0:
        continue; # 홀수면 재실행 (위로)
    print (a);
```
