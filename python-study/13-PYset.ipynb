```python
#set
#집합을 의미한다.
#딕셔너리와 같이 순서가 없습니다
#****값의 중복을 허락하지 않는다

가족 = {"요뚜비","조먼지","조하루"};

#순서도 없고 출력하면 요소가 다르게 나오는것도 특징이다
print(가족);

세트 = set({"가","나","다"});
print(세트);
#데이터 삽입
세트.add("라");
print(세트);

#데이터 삽입2
세트.update("마바");
print(세트); #차이를 느껴봐

#데이터 삭제
#1 remove 없으면 error
세트.remove("바");
print(세트);

#2.discard 없어도 실행
세트.discard("바");

#3.데이터 삭제하고 반환
print(세트.pop());

# 합집합 교집합 차집합


고양이 = {"조먼지","조하루","모래","포포"};
가족 = {"요뚜비","조하루","조먼지"};
#합집합
print(고양이|가족);
set.union(고양이,가족);

#교집합
print(고양이&가족);
set.intersection(고양이,가족);

#차집합
print(고양이-가족);
set.difference(고양이,가족);





```
