```python
#slicing (슬라이싱)
로또 = [3,6,17,24,36,44];
#변수 인덱스 1~2까지만 출력
print("인덱스 : ",로또[1:3]);

#마지막 위치 출력
print("마지막 : ",로또[-1]);

#반대로
print("반대로 : ",로또[::-1]);

#로또 36번만 출력
print("36번만 : ", 36 in 로또);

#로또 와 로또2를 합쳐
로또2 = [4,27,28,33,37,41];
print("두개 : ",로또 + 로또2);

print("");

#range를 이용해서 1부터 10사이에 짝수만 들어있는 리스트를 만들어라
짝수 = list(range(2,12,2));
print("짝수 :",짝수);

#짝수 리스트에있는 값을 2배 늘려 다시 짝수에 담아라
짝수 = 짝수*2;
print("2배 : ",짝수);

#짝수리스트의 요수 갯수를 출력해라
print("길이 : ",len(짝수));

#마지막 수 출력
print("마지막 : ",짝수[len(짝수)-1]);
```

```python
#리스트 값 넣고 빼고 바꾸고

저메추 = ["막창","곱창전골","국밥","치킨","부찌","된찌","김찌"];

#값넣기 append():마지막에 값을 넣음 insert():원하는곳에 넣음
저메추.append("닭발");
print(저메추);

저메추.insert(2,"썹웨이");
print(저메추);

#값뺴기
del 저메추[-1];
print(저메추);

#값 바꾸기
저메추[-1] = "닭발";
print(저메추);
```

```python
print(저메추);

저메추.reverse();
print(저메추);

저메추.sort();
print(저메추);
```
