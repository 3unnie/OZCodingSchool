```python
팥가격 = 2000;
슈크림가격 = 2500;
잡채가격 = 3000;

팥갯수 = 0;
슈크림갯수 = 0;
잡채갯수 = 0;

팥총가격 = 0;
슈크림총가격 = 0;
잡채총가격 = 0;

while True:

    맛 = input("무슨 맛을 고르실껀가요 [팥/슈크림/잡채]");

    if 맛 == "팥":
        print("팥을 선택하셧습니다");
        print("가격 : 2000 원");

        팥갯수 += int(input("몇개를 구매하시겟습니까?"));

        if 팥갯수 > 0:
            print(팥갯수,"개 팥 붕어빵 주문");
            팥총가격 = 팥가격 * 팥갯수;

            질문 = input("더 주문 하시겠습니까? (예/아니오)");
            if 질문 =="예":
                continue;
            else:
                print("주문을 종료하겠습니다.")
                break;

    elif 맛 == "슈크림":
        print("슈크림을 선택하셧습니다");
        print("가격 : 2500 원");

        슈크림갯수 += int(input("몇개를 구매하시겟습니까?"));

        if 슈크림갯수 > 0:
            print(슈크림갯수,"개 슈크림 붕어빵 주문");

            슈크림총가격 = 슈크림가격 * 슈크림갯수;

            질문 = input("더 주문 하시겠습니까? (예/아니오)");
            if 질문 =="예":
                continue;
            else:
                print("주문을 종료하겠습니다.")
                break;

    elif 맛 == "잡채":
        print("잡채을 선택하셧습니다");
        print("가격 : 3000 원");

        잡채갯수 += int(input("몇개를 구매하시겟습니까?"));

        if 잡채갯수 > 0:
            print(잡채갯수,"개 잡채 붕어빵 주문");

            잡채총가격 = 잡채가격 * 잡채갯수;

            질문 = input("더 주문 하시겠습니까? (예/아니오)");
            if 질문 =="예":
                continue;
            else:
                print("주문을 종료하겠습니다.")
                break;

    else:
        결정 = input("구매를 원하세요(예/아니오)");
        if 결정 == "예":
            print("맛을 정확히 입력해 주세요");
            continue;
        else:
            print("안녕히 가세요")
            break;

총가격 = 팥총가격 + 슈크림총가격 + 잡채총가격;

if 총가격 > 0:
    print();
    print("[주문내역]");
    print("밭 붕어빵 :", 팥갯수,"개");
    print("슈크림 붕어빵 :", 슈크림갯수,"개");
    print("잡채 붕어빵 :", 잡채갯수,"개");
    print("총", 총가격,"원 입니다.");
    print("감샤링~")


```
