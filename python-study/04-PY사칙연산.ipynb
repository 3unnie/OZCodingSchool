##사칙연산
더하기 : +

뺴기 : -  

곱하기 : *  

나누기 :
*   /(우리가 알고있는 나누기)
*   // (몫을 반환)
*   %  (나머지를 반환)

거듭제곱 : **

```python
#더하기
print(3+1)
```

```python
#뺴기
print(3-1)
```

```python
#곱하기
print(3*2)
```

```python
#나누기
print(6/3)
print(12//5)
print(12//5)
```

```python
#거듭제곱
print(3**3)
```

내장 함수 : divmod (몫과 나머지를 동시에 구해줌)
> 자주 사용하는 기능을  미리 만들어 놓은것을 내장함수





```python
print(divmod(7,3))
```
