```python
##입력과 출력

#입력 input
#출력 print

#함수와 메서드
# 함수 : 어디에 속해 있지 않고 단독 모듈이라 함수를 그냥 호출하여 사용
# 메서드 : 함수의 클래스 안에 속해있고 클래스의 멤버 변수들을 이용해서 구현된 것


인사 = input("반가워"); #input은 str으로 들어온다
print(인사 + "나도");

#형변환
#int
정수 = input("정수를 입력해봐");
정수 = int(정수);
print (정수, type(정수));

#float
실수 = input("실수를 입력해봐");
실수 = float(실수);
print (실수, type(실수));

#그외 형변환 가능종류 [tuple() list() dict() set()]
```

```python
넌센스, 정답 = input().split("/");

print("넌센스퀴즈 : " + 넌센스);
print("정답 : "+ 정답);
```

```python
#map
num1, num2, num3 = map(int ,input().split("-"));

print(num1,num2,num3);
```
