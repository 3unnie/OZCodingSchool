```python
#dict

#list tuple str range
#인덱스 내용에 연관성이 없다

#but dict는 여러개의 값이 일렬로 정렬되면서 값끼리의 연관성이 존재

고양이1 = {"이름":"조먼지", "성별":"남", "나이":0.9, "성격" : "찐따"};
고양이2 = {"이름":"조하루", "성별":"남", "나이":0.2, "성격" : "마마보이"};

#{키 : 값} (키값은 중복이 되지않아요~)
#{key1 : value1}

print(고양이1["이름"]);
```

```python
#값을 넣어 만드는법
값넣딕 = dict(이름="조먼지", 성별="남");
print(값넣딕);

#기호 주의~
```

```python
#zip : 병렬정리를 도와준다

집 = dict(zip(["이름","성별"],["조하루","남"]));
print(집);
```

```python
print(고양이1.keys());
print(고양이2.values());
```

```python
print(고양이2.items());

리스트화 = list(고양이1.items());
print(리스트화); #2차원 리스트로 변신

print(리스트화[0][1]);
```

```python
#딕셔너리 값추가
고양이1.update({"몇째": 1, "생년월일" : "6월5일"});
print(고양이1);
```
